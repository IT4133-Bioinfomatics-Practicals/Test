```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('dna_samples.csv')

def valid_dna(seq):
    for i in seq:
        if i not in 'ATGC':
            return False
        return True
    
sample_ids=[]
lengths=[]
A_list=[]
T_list=[]
G_list=[]
C_list=[]
GC_Content=[]

for i in range(len(df)):
    sid=df.loc[i,'SampleID']
    seq=df.loc[i,'DNA_Sequence'].upper()

    if not valid_dna(seq):
        print("Invalid DNA ", sid)
        continue

    length=len(seq)
    A = seq.count("A")
    T = seq.count("T")
    G = seq.count("G")
    C = seq.count("C")
    GC = ((G+C)/length) * 100

    sample_ids.append(sid)
    lengths.append(length)
    A_list.append(A)
    T_list.append(T)
    G_list.append(G)
    C_list.append(C)
    GC_Content.append(GC)

summary = pd.DataFrame({
    "Sample ID":sample_ids,
    "Length":lengths,
    "A":A_list,
    "T":T_list,
    "G":G_list,
    "C":C_list,
    "GC":GC_Content
})

print(summary)
```

```python
plt.bar(summary['Sample ID'], summary['GC'])
plt.show()
```
