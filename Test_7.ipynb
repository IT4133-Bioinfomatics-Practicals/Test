```python
# complenent

dna = "ATGGCGTAGTGA"

def complement_dna(seq):
    complement = {'A':'T', 'T':'A', 'G':'C', 'C':'G'}
    return "".join(complement[base] for base in seq)

complement1 = complement_dna(dna)
print(complement1)
```

```python
# reverce

reverse_dna = "".join(reversed(dna))
print(reverse_dna)
```

```python
# Score DNA

s1 = "AGC-T"
s2 = "AGCAT"

def score_dna(seq1, seq2):
    matches=2
    mismatch=-1
    gap=-2
    score=0

    for a,b in zip(seq1, seq2):
        if a=='-' or b=='-':
            score += gap
        elif a == b:
            score += matches
        else:
            score += mismatch
    return score

print(score_dna(s1, s2))
```

```python
# global sequence (nw algorithm)

s1 = "AGC-T"
s2 = "AGCAT"

def nw_score(seq1, seq2):
    match = 2
    mismatch = -1
    gap = -2

    m = len(seq1)
    n = len(seq2)

    dp = [[0]*(n+1) for _ in range(m+1)]

    for i in range(m+1):
        dp[i][0] = i * gap
    for j in range(n+1):
        dp[0][j] = j * gap

    for i in range(1, m+1):
        for j in range(1, m+1):
            score = match if seq1[i-1] == seq2[j-1] else mismatch

            dp[i][j] = max(
                dp[i-1][j-1] + score,
                dp[i-1][j] + gap,
                dp[i][j-1] + gap
            )
    return dp[m][n]

print("Needleman-Wunsch Score: ", nw_score(s1, s2))
```
