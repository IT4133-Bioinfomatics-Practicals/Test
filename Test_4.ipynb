```python
# Define DNA sequence for analysis
dna = "ATGGCGTAGTGA"

# Function to generate complementary DNA strand
def complement_dna(seq):
    # Dictionary mapping each nucleotide to its complement
    complement = {'A':'T', 'T':'A', 'G':'C', 'C':'G'}
    return "".join(complement[base] for base in seq)

# Generate and display the complementary DNA sequence
complement1 = complement_dna(dna)
print(complement1)
```

```python
# Reverse the DNA sequence (5' to 3' becomes 3' to 5')
reverse_dna = "".join(reversed(dna))
print(reverse_dna)
```

```python
# Define two DNA sequences for alignment scoring (with gap in sequence 1)
s1 = "AGC-T"
s2 = "AGCAT"

# Function to calculate alignment score using simple scoring scheme
def score_dna(seq1, seq2):
    # Define scoring parameters
    match=2        # Score for matching bases
    mismatch=-1    # Penalty for mismatched bases
    gap=-2         # Penalty for gaps
    score=0

    # Compare sequences position by position
    for a,b in zip(seq1, seq2):
        if a=='-' or b=='-':
            score += gap          # Apply gap penalty
        elif a == b:
            score += match        # Add match score
        else:
            score += mismatch     # Apply mismatch penalty
    return score

# Calculate and display alignment score
print(score_dna(s1, s2))
```

```python
# Define sequences for global alignment
s1 = "AGC-T"
s2 = "AGCAT"

# Function to calculate optimal alignment score using Needleman-Wunsch algorithm
def nw_score(seq1, seq2):
    # Define scoring parameters
    match = 2       # Score for matching bases
    mismatch = -1   # Penalty for mismatched bases
    gap = -2        # Penalty for gaps

    # Get sequence lengths
    m = len(seq1)
    n = len(seq2)

    # Initialize dynamic programming matrix
    dp = [[0]*(n+1) for _ in range(m+1)]

    # Initialize first column (gaps in seq2)
    for i in range(m+1):
        dp[i][0] = i * gap
    # Initialize first row (gaps in seq1)
    for j in range(n+1):
        dp[0][j] = j * gap

    # Fill the dynamic programming matrix
    for i in range(1, m+1):
        for j in range(1, n+1):
            # Calculate match/mismatch score
            score = match if seq1[i-1] == seq2[j-1] else mismatch
            
            # Choose maximum score from three possible operations
            dp[i][j] = max(
                dp[i-1][j-1] + score,  # Diagonal: match/mismatch
                dp[i-1][j] + gap,       # Up: gap in seq2
                dp[i][j-1] + gap        # Left: gap in seq1
            )
    
    # Return the optimal alignment score
    return dp[m][n]

# Calculate and display Needleman-Wunsch alignment score
print("Needleman-Wunsch Score: ", nw_score(s1, s2))
```
