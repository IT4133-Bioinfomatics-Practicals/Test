```python
import numpy as np

L = ['A', 'B', 'C', 'D']

arr = np.array([[0,5,9,9],
                [5,0,10,10],
                [9,10,0,8],
                [9,10,8,0]
                ], float)

while len(L) > 1:
    i,j = np.unravel_index(np.argmin(arr+np.eye(len(arr))*1e10), arr.shape)

    print(L[i], '+', L[j])

    new = (arr[i] + arr[j]) / 2
    arr = np.delete(arr, [i,j], 0)
    arr = np.delete(arr, [i,j], 1)
    arr = np.c_[arr, new[:-2]]
    arr = np.r_[arr, [np.append(new[:-2], 0)]]
    L += [L[i] + L[j]]
    L.pop(max(i,j))
    L.pop(min(i,j))
```
