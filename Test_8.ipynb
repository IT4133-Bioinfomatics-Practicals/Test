```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('dea_data.csv')
df.head()
```

```python
genes = df['Gene']

df.describe()
```

```python
# Identify Columns

control_cols = [c for c in df.columns if 'Control' in c]
treatment_cols = [c for c in df.columns if 'Treatment' in c]

display("Control Colums: ", df[control_cols])
display("Treatment Columns: ", df[treatment_cols])
```

```python
# Calculate mean

df['Control Mean'] = df[control_cols].mean(axis=1)
df['Treatment Mean'] = df[treatment_cols].mean(axis=1)

display("Control Mean: ", df['Control Mean'])
display("Treatment Mean: ", df['Treatment Mean'])
```

```python
# Fold Change

epsilon = 1e-6
df['Fold_Change'] = (df['Treatment Mean'] + epsilon) / (df['Control Mean'] + epsilon)
df['log2FC'] = np.log2(df['Fold_Change'])

display("Fold Change: ", df['Fold_Change'])
display("log2 Fold Change: ", df['log2FC'])
```

```python
# Classify Gene

def classify(fc):
    if fc > 1:
        return 'Up_regulated'
    elif fc <-1:
        return 'Down_regulated'
    else:
        return 'No_change'
    
status = df['Fold_Change'].apply(classify)
```

```python
# Create result table

dea_results = pd.DataFrame({
    'Gene': genes,
    'Control Mean': df['Control Mean'],
    'Treatment Mean': df['Treatment Mean'],
    'Fold Change': df['Fold_Change'],
    'log2 Fold Change': df['log2FC'],
    'Status': status
})

display(dea_results)
```

```python
# save results

dea_results.to_csv('dea_results.csv', index=False)
```
