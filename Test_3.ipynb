```python
import pandas as pd
import numpy as np

df = pd.read_csv('dea_data.csv')
df.describe()

control_cols = [c for c in df.columns if "Control" in c]
treatment_cols = [c for c in df.columns if "Treatment" in c]

df['Control Mean'] = df[control_cols].mean(axis=1)
df['Treatment Mean'] = df[treatment_cols].mean(axis=1)

epsilon = 1e-6
df['Fold_Change'] = (df['Treatment Mean'] + epsilon) / (df['Control Mean'] + epsilon)
df['log2FC'] = np.log2(df['Fold_Change'])

def classify(fc):
    if fc >= 1:
        status = "Up_regulated"
    elif fc <= 1:
        status = "Down_regulated"
    else:
        status = "Not change"
df['status'] = df['Fold_Change'].apply(classify)

final = df[['Gene', 'Control Mean', 'Treatment Mean', 'Fold_Change', 'log2FC', 'status']]
print(final)
```

```python
final.to_csv('New1.csv', index=False)
```
